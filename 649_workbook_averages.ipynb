```python
import pandas as pd
```

```python
df = pd.read_csv("presidents_general.csv")
occupations = pd.read_csv('occupations.csv')
```

```python
# Most common birth state for each era of presidents
print(df.groupby('era')['Birth State'].agg(pd.Series.mode))
```

```python
# Most common death state for each era of presidents
print(df.groupby('era')['Death State'].agg(pd.Series.mode))
```

```python
# Most common university attended for each era of presidents
print(df.groupby('era')['University'].agg(pd.Series.mode))
```

```python
# Most common highest degree held for each era of presidents
print(df.groupby('era')['Highest Degree Held'].agg(pd.Series.mode))
```

```python
# Average number of countries visited for each era of presidents
print(df.groupby('era')[
      '# Countries Visited as President'].mean())
```

```python
# Average age at taking office for each era of presidents
print(df.groupby('era')[
      'Age Taking Office'].mean())
```

```python
# Average height for each era of presidents
print(df.groupby('era')[
      'Height'].mean())
```

```python
# Average number of children for each era of presidents
print(df.groupby('era')[
      'Children'].mean())
```

```python
# Most common religion for each era of presidents
print(df.groupby('era')[
      'Religion'].agg(pd.Series.mode))
```

```python
# Most common occupation for each era of presidents
print(occupations.groupby('era')[
      'occupation_group'].agg(pd.Series.mode))
```

```python
# Most common party for each era of presidents
print(df.groupby('era')[
      'Party'].agg(pd.Series.mode))
```

```python
# Most common previous office held for each era of presidents
print(df.groupby('era')[
      'previous_office_group'].agg(pd.Series.mode))
```

```python
# Most common highest military position held for each era of presidents
print(df.groupby('era')[
      'military_level'].agg(pd.Series.mode))
```
